# Importing Libraries 

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

# List of Stop Words

```python
stopwords = list(STOP_WORDS)
```

# Text

```python
text = """"Mahmoud Darwish is a Palestinian poet and member of the Palestinian National Council of the Palestine Liberation Committee. He has poetry collections full of modernist content. He was born in 1941 in the village of Al-Birwa, a Palestinian village located in the Galilee near the coast of Acre, where his family owned land. The family went out with the Palestinian refugees in 1948 to Lebanon, then returned infiltrated in 1949 after the armistice was signed, only to find the village destroyed and a moshav (an Israeli agricultural village) “Ahihud” built on its lands. Kibbutz Yas'ur lived with his family in the new village.
After sending his secondary education in a high school in high school, he joined the Israeli Communist Party, the new party in the party, and the new one who edited it, edited it, edited it, edited Al-Fajr newspaper, which was published by Mapam.
He served as president of the General Union of Palestinian Writers and Writers and edited Al-Karmel magazine. His residence was in Paris before his return to his homeland, as he entered Palestine with a permit to visit his mother. While he was there, some Arab and Jewish members of the Israeli Knesset made a proposal to allow him to stay, and he was allowed to do so.
Critics divided the stages of Darwish's poetry into several sections that combine the poet's relationship with his homeland, his cause "the Palestinian cause", exile and leaving home, all in light of his relationship with the self. Critic Muhammad Fikri Al-Jazzar divided Darwish's poetry into three sections: The first stage, which is the stage of his presence in the homeland, which includes the beginnings of the poet's formation and awareness of the cause of his homeland and the formation of belonging to this homeland under the occupation."""
```

```python
# Object to process the text 
nlp = spacy.load('en_core_web_sm')
```

```python
# Segminting the text into words, punctuation and so on
doc = nlp(text)
```

```python
#each doc consists of individual tokens
# List of tokens
tokens = [token.text for token in doc]
print(tokens)
```

```python
# Removing punctuation & Adding '\n'
punctuation = punctuation + '\n'
punctuation
```

```python
# Counting the number of occurrences of all the distinct words which are not punctuations or stopwords 
word_frequencies = {}
for word in doc:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:
            if word.text not in word_frequencies.keys():
                word_frequencies[word.text] = 1
            else:
                word_frequencies[word.text] += 1
                
print(word_frequencies)
```

```python
# Get the max_frequency: the most repeated word
max_frequency = max(word_frequencies.values())
max_frequency
```

# Normalizing the Frequencies

```python
# We will divide each frequency value in word_frequencies with the max_frequency to normalize the frequencies.
for word in word_frequencies.keys():
    word_frequencies[word] = word_frequencies[word] / max_frequency

print(word_frequencies)
```

```python
# Dividing the text into Sentences
sentence_tokens = [sent for sent in doc.sents]
print(sentence_tokens)
```

# Calculating the Sentence Scores

```python
# The sentence score for a particular sentence is the sum of the normalized frequencies of the words in that sentence. 
sentence_scores = {}
for sent in sentence_tokens:
    for word in sent:
        if word.text.lower() in word_frequencies.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent] = word_frequencies[word.text.lower()]
            else:
                sentence_scores[sent] += word_frequencies[word.text.lower()]
                
sentence_scores
```

```python
from heapq import nlargest
```

```python
# The length of the summary is 30% of the original length
select_length = int(len(sentence_tokens) * 0.3)
select_length
```

```python
# Returning a list with the select_length largest elements 
summary = nlargest(select_length, sentence_scores, key = sentence_scores.get)
summary
```

```python
# Combining these sentences to the final summary 
final_summary = [word.text for word in summary]
summary = ' '.join(final_summary)
```

# Displying the Original Text & The Generated Summary

```python
# Displaying the original text 
print(text)
```

```python
# Displaying the generated summary
print(summary)
```
